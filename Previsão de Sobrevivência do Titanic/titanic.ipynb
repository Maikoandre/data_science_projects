```python
import polars as pl
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['figure.dpi'] = 100

%matplotlib inline
```

```python
df = pl.read_csv('dados/titanic.csv')
df.head()
```

```python
df.schema
```

```python
df.describe()
```

```python
missing = (
    df.select(pl.all().null_count())
    .transpose(include_header=True, header_name='column', column_names=['Missing Count'])
    .with_columns(
        ((pl.col("Missing Count") / df.height) * 100).round(2).alias('Missing %')
    ).filter(pl.col('Missing Count') > 0).sort('Missing %', descending=True)
)
    
missing
```
